# **Violinplot + Swarmplot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head(2)
```

```python
sns.violinplot(x='tip', y='day', data=df)
sns.swarmplot(x='tip', y='day', data=df)
```
