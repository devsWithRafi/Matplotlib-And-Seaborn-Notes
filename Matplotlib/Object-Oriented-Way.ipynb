```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

    Matplotlib allows the aspact retio, DPI and figure size to be spacified when the fegure object is created.
    You can use the figsize and DPI keyword arguments.

    Figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

    DPI = Dots-Per-Inch

```python
x = np.linspace(0, 5, 11)
print(x)

y = x**3
print(y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes1.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes1.plot(x**2,x**2)
axes2 = fig.add_axes([0.4, 0.2, 0.4, 0.4])
axes2.plot(x,y)
```

**Using "figzise" and "dpi"**

```python
fig = plt.figure(figsize=(10,5), dpi=100)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes1.plot(x,y)
```
